```python
df1[df1.Age>40]
#you’re doing boolean indexing — you give pandas a Series of True/False values (one for each row), and it returns only the rows where the value is True.


#Why can’t we just write inside Age > 40 directly?

#If you wrote just:

#df1[Age > 40]


#pandas wouldn’t know what Age is.

#Age is not a standalone Python variable — it’s a column inside df1.

#To tell pandas which column you’re talking about, you must access it first:
```
